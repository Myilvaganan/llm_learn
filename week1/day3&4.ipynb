## Tokenizing with code


```python
import tiktoken

token_model = tiktoken.encoding_for_model("gpt-4.1-mini")

tokens = token_model.encode("Hi my name is Ed and I like banoffee pie")
```

```python
tokens
```

```python
for token in tokens:
    print(token, "=",token_model.decode([token]))
```

```python
from openai import OpenAI

openai = OpenAI()
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Hi! I'm Ed!"},
    {"role": "assistant", "content": "Hi Ed! How can I assist you today?"},
    {"role": "user", "content": "What's my name?"}
    ]
```

```python
response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
response.choices[0].message.content
```
